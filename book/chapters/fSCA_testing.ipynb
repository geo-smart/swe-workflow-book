```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Sample DataFrame creation for demonstration
# Replace this part with your actual DataFrame
df = pd.DataFrame({
    'date': pd.date_range(start='2023-01-01', periods=10),
    'SCA_percentage': [50, 55, 60, 65, 70, 65, 60, 55, 50, 45]
})

# Ensure 'date' column is in datetime format for plotting
df['date'] = pd.to_datetime(df['date'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['SCA_percentage'], marker='o', linestyle='-', color='b')
plt.title('Snow Cover Area (SCA) Percentage Over Time')
plt.xlabel('Date')
plt.ylabel('SCA Percentage (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
